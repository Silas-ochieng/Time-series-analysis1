# Data Cleaning and Preprocessing:

# outlier detection and removal

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv((r"C:\Users\san\Desktop\Datasets\GOOG.csv"))
data = data[['Volume','Close']]
print(data.head())
print(data)
data.shape
data.sample(5)
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.histplot(data['Volume'], kde=True)
plt.subplot(1,2,2)
sns.histplot(data['Close'], kde=True)
plt.show()
data['Volume'].skew()
data['Close'].skew()
print("Mean value of Close",data['Close'].mean())
print("Standard deviation of Close",data['Close'].std())
print("Minimum value of Close",data['Close'].min())
print("Maximum value of Close",data['Close'].max())
print("Upper limit",data['Close'].mean() + 3*data['Close'].std())
print("Lower limit",data['Close'].mean() - 3*data['Close'].std())
data[(data['Close'] > 946.2225996174238) | (data['Close'] < 249.3714992397188)]
```
